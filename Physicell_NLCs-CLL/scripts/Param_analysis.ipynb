```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##### Load experimental data

```python
experimental = np.loadtxt('../../Netlogo_NLCs-CLL/filtered_fused_9patients.csv', delimiter=",", skiprows=1)
viability_exp = experimental[:,1]
concentration_exp = experimental[:,2]
time_points = experimental[:,0]
```

##### Load output from simulation

```python
viability_data_node1 = pd.read_csv('../data_output/Parameter_exploration/results/viability_exploration_1.csv', delimiter=",")
viability_data_node2 = pd.read_csv('../data_output/Parameter_exploration/results/viability_exploration_2.csv', delimiter=",")
viability_data_node3 = pd.read_csv('../data_output/Parameter_exploration/results/viability_exploration_3.csv', delimiter=",")
viability_data = pd.concat([viability_data_node1, viability_data_node2, viability_data_node3], axis = 1) 

concentration_data_node1 = pd.read_csv('../data_output/Parameter_exploration/results/concentration_exploration_1.csv', delimiter=",")
concentration_data_node2 = pd.read_csv('../data_output/Parameter_exploration/results/concentration_exploration_2.csv', delimiter=",")
concentration_data_node3 = pd.read_csv('../data_output/Parameter_exploration/results/concentration_exploration_3.csv', delimiter=",")
concentration_data = pd.concat([concentration_data_node1, concentration_data_node2, concentration_data_node3], axis=1)

vals = list(np.round(np.linspace(0, 1, 20), 2))
len(vals)

viability_data.shape[1]

input = {'uptake_rate_cancer': 1.0, 'speed_cancer': 1.0, 'transformation_rate_cancer': 5e-5,
                  'speed_monocytes':1.0, 'dead_phagocytosis_rate_monocytes':25e-2, 'speed_macrophages':1.0,
                  'dead_phagocytosis_rate_macrophages':92e-2, 'secretion_rate_NLCs':1.0, 'speed_NLCs':1.0,
                  'dead_phagocytosis_rate_NLCs':4e-2, 'death_rate_apoptotic':3e-3, 'secretion_rate_apoptotic': 1.0}

param_names = list(input.keys())
```

Exploration with new values

```python
vals = list([5e-7, 5e-5, 5e-3, 5e-1])
viability_data = pd.read_csv('../data_output/Parameter_exploration/results/viability_new.csv', delimiter=",")
concentration_data = pd.read_csv('../data_output/Parameter_exploration/results/concentration_new.csv', delimiter=",")
input = {'uptake_rate_cancer': 1.0, 'speed_cancer': 1.0, 'transformation_rate_cancer': 5e-5,
                  'speed_monocytes':1.0, 'dead_phagocytosis_rate_monocytes':25e-2, 'speed_macrophages':1.0,
                  'dead_phagocytosis_rate_macrophages':92e-2, 'secretion_rate_NLCs':1.0, 'speed_NLCs':1.0,
                  'dead_phagocytosis_rate_NLCs':4e-2, 'death_rate_apoptotic':3e-3, 'secretion_rate_apoptotic': 1.0}

param_names = list(input.keys())
```

```python
contador = 0
for k in range(0, viability_data.shape[1], len(vals)):
    
    viability = viability_data.iloc[:,k:k+ len(vals)]
    concentration = concentration_data.iloc[:,k:k+ len(vals)]

    ###Viability
    for i in range(viability.shape[1]): # Iterate over each column of the matrix
        plt.plot(time_points, viability.iloc[:, i], label=f'Value = {vals[i]}')

    plt.plot(time_points, viability_exp, label='Experimental', color = 'black')
    plt.xlabel('Time')
    plt.ylabel('Viability')
    plt.title(f'Viability: {param_names[contador]}')
    plt.legend(loc='upper right', bbox_to_anchor=(1.35, 1))

    plt.grid(True)
    plt.show()

    ###Concentration
    for i in range(concentration.shape[1]):
        plt.plot(time_points, concentration.iloc[:, i], label=f'Value = {vals[i]}')

    plt.plot(time_points, concentration_exp, label='Experimental', color = 'black')
    plt.xlabel('Time')
    plt.ylabel('Concentration')
    plt.title(f'Concentration: {param_names[contador]}')
    plt.legend(loc='upper right', bbox_to_anchor=(1.35, 1))

    plt.grid(True)
    plt.show()
    
    contador = contador + 1
```
